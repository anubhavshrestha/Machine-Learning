## 1.1. Create a class
* Write a class named `Employee` that holds the following data about an employee in attributes: 'name', 'ID number', 'department', and 'job title'.

````python
class Employee:
  def __init__(self, name, id, dept, job_title):
    self.name = name
    self.id = id
    self.dept = dept
    self.job_title = job_title
  def __str__(self):
    return(f"Employee Name: {self.name} \nID Number: {self.id} \nDepartment: {self.dept} \nJob Title: {self.job_title}")
````

## 3. Generator

A pair of twin primes is a pair of consecutive prime numbers that differ by two. The first three such pairs are (3, 5), (5, 7), and (11,13).

Write a generator function that yields tuples representing these pairs one after another. Call the next method on this generator function 10 times and print out the first ten such pairs.

````python
def is_prime(n):
  if n <= 1:return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
        return False
  return True

def generator(n):
  """n is the number of consecutive primes we want to generate"""
  num = 2
  total_pairs = n
  pairs_count = 0
  while pairs_count < n:
    if (is_prime(num) & is_prime(num+2)):
      yield (num, num+2)
    num += 1

iterator = generator(20)    # generate 20 pairs of consecutive prime that differs by two

for i in range(10):
  print(next(iterator))
````

Then set the prices at Charlie's at \$1 more than Beth's. Print the resultant dataframe.

````python
df_restaurant = pd.DataFrame({'restaurant_name': ["Amy's Pizza", "Beth's Pizza", "Charlie's Pizza"]})
df_slice = pd.DataFrame({'slice_type': ["plain", "mushroom", "anchovy"]})

df_cartesian = pd.merge(how = "cross", left = df_restaurant, right = df_slice)

df_cartesian['price'] = 0  # Initialize price column

price_dict = {
    "Amy's Pizza": {'plain': 1, 'mushroom': 1.25, 'anchovy': 1.5},
    "Beth's Pizza": {'plain': 2, 'mushroom': 2.25, 'anchovy': 2.5},
    "Charlie's Pizza": {'plain': 3, 'mushroom': 3.25, 'anchovy': 3.5}
}

#initializing the price of everyone according to slice types
df_cartesian.loc[df_cartesian['slice_type'] == 'plain', 'price'] = 1
df_cartesian.loc[df_cartesian['slice_type'] == 'mushroom', 'price'] = 1.25
df_cartesian.loc[df_cartesian['slice_type'] == 'anchovy', 'price'] = 1.5

#Adding one to Beth and Charlie
df_cartesian.loc[(df_cartesian['restaurant_name'] == "Beth's Pizza") | (df_cartesian['restaurant_name'] == "Charlie's Pizza"), 'price'] +=1

#Adding one more to Charlie
df_cartesian.loc[df_cartesian['restaurant_name'] == "Charlie's Pizza", 'price'] +=1

print(df_cartesian)
````
